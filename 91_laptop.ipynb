```python
with open('webLaptop91.html','r',encoding = 'utf-8') as f:
  html = f.read()
```

```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html,'lxml')
```

```python
containers = soup.find_all('div',{'class':'grid-inner'})
```

```python
len(containers)
```

```python
len(containers)
```

```python
name =[]
for i in containers:
  j = i.find('a',{'class':'hover_blue_link name gaclick'})
  if j:
    name.append(j.text.strip())

  else:
    name.append(np.nan)
```

```python
len(name)
```

```python
print(name)
```

```python
spec_score = []
for i in containers:
  j = i.find('div',{'class':'rating_box_new_list spec_score_translated verygood pos_abs various3 fzancybox.iframe gaclick'})
  if j:

    spec_score.append(j['href'].split('&score=')[1])
  else:
    spec_score.append(np.nan)
```

```python
from pprint import pprint
```

```python
img = []
name_c = []
for i in containers:
  j = i.find('img',{'class':'fimage gaclick fndr_lptp'})
  if j:
    img.append(j['src'].strip())
    name_c.append(j['alt'].strip())
  else:
    img.append(np.nan)
    name_c.append(np.nan)
```

```python
price = []
for i in containers:
  j = i.find('div',{'class':'pro_grid_price'})
  if j:
    price.append(j.text.split(' ')[1].replace(',',''))

  else:
    price.append(np.nan)
```

```python
no_rates = []
for i in containers:
  j = i.find('p')
  if j:
    no_rates.append(j.text.split(' ')[0].replace(',',''))
  else:
    no_rates.append(np.nan)

```

```python
rating_100 = []
for i in containers:
  j = i.find('div',{'class':'rating'})
  if j:
    rating_100.append(j['style'].split( )[1].replace('%;',''))
  else:
    rating_100.append(np.nan)
```

```python
released = []
for i in containers:
  j = i.find('div',{'class':'pro_grid_date'})
  if j:
    released.append(j.text.split(':')[1].strip())
  else:
    released.append(np.nan)
```

```python
system_name = []
for i in containers:
  j = i.find('div',{'class':'featuresblock'})
  if j:
    system_name.append(j.text.strip())
  else:
    system_name.append(np.nan)
```

```python
len(system_name)
len(name)
```

```python
processor = []
windows = []
storage = []
display = []
for i in containers:
  j = i.find_all('div',{'class':'featuresblock',})
  if j:
    specification = [jin['title'] for jin in j]
    windows.append(specification[0])
    processor.append(specification[1])
    storage.append(specification[2])
    display.append(specification[3])
  else:
    windows.append(np.nan)
    processor.append(np.nan)
    storage.append(np.nan)
    display.append(np.nan)
```

```python
Wifi = []
bluetooth = []
camera = []
slim = []
hdmi = []
usb = []

for i in containers:
  j = i.find('div',{'class':'tooltip finder_icon icn_wifi'})
  if j:
    Wifi.append(j['data-title'])
  else:
    Wifi.append(np.nan)

```

```python
for i in containers:
  j = i.find('div',{'class':'tooltip finder_icon icn_bluetooth'})
  if j:
    bluetooth.append(j['data-title'])
  else:
    bluetooth.append(np.nan)
```

```python
for i in containers:
  j = i.find('div',{'class':'tooltip finder_icon icn_camera'})
  if j:
    camera.append(j['data-title'])
  else:
    camera.append(np.nan)
```

```python
for i in containers:
  j = i.find('div',{'class':'tooltip finder_icon icn_slim'})
  if j:
    slim.append(j['data-title'])
  else:
    slim.append(np.nan)
```

```python
for i in containers:
  j = i.find('div',{'class':'tooltip finder_icon icn_HDMIPorts'})
  if j:
    hdmi.append(j['data-title'])
  else:
    hdmi.append(np.nan)
```

```python
for i in containers:
  j = i.find('div',{'class':'tooltip finder_icon icn_usbports'})
  if j:
    usb.append(j['data-title'])
  else:
    usb.append(np.nan)
```

```python
df = pd.DataFrame({
    'name':name,
  'name_c':name_c,
  'img':img,
  'price':price,
    'no_rates':no_rates,
  'rating_100':rating_100,
  'released':released,
  'system_name':system_name,
    'processor':processor,
  'windows':windows,
  'storage':storage,
  'display':display,
    'Wifi':Wifi,
  'bluetooth':bluetooth,
  'camera':camera,
  'slim':slim,
  'hdmi':hdmi,
  'usb':usb,
  'spec_score':spec_score

})
```
