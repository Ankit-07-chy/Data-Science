```python
# application
import numpy as np
import pandas as pd

np.random.seed(23)

mu_vec1 = np.array([0,0,0])
cov_mat1 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class1_sample = np.random.multivariate_normal(mu_vec1, cov_mat1, 20)
print('class1_sample : \n',class1_sample)

df= pd.DataFrame(class1_sample,columns=['f1','f2','f3'])
df['target'] = 1
print(df.head())

mu_vec2 = np.array([1,1,1])
cov_mat2  = np.array([[1,0,0],[0,1,0],[0,0,1]])
class2_sample = np.random.multivariate_normal(mu_vec2, cov_mat2, 20)
print('class2_sample : \n',class2_sample)

df2= pd.DataFrame(class2_sample,columns=['f1','f2','f3'])
df2['target'] = 0
print(df2.head())

df = pd.concat([df, df2], ignore_index=True)
df = df.sample(40)
```

```python
import plotly.express as px
#y_train_trf = y_train.astype(str)
fig = px.scatter_3d(df, x='f1', y='f2', z='f3',
              color='target')
fig.update_traces(marker_size=5)
fig.show()
```

```python
# step 1- spply standard scaling
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
df.iloc[:,0:3] = scalar.fit_transform(df.iloc[:,0:3])
df.head()
```

```python
covariance_matrix = np.cov(df.iloc[:,0:3].T)
print(covariance_matrix)
```

```python
eigen_values , eigen_vectors = np.linalg.eig(covariance_matrix)
print(eigen_values)
print(eigen_vectors)
```
