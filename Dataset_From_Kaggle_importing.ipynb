```python
# lets run on events
events.head()
```

```python
events['Season'] = events['edition'].str.split(' ').str.get(1)
```

```python
# lets analysize for summer season only
events = events[events['Season'] == 'Summer']
```

```python
events = events.merge(profiles, left_on='country_noc',right_on='noc').drop(columns=['country_noc'])
```

```python
events['country'].value_counts()
```

```python
events.isnull().sum()
```

```python
events.duplicated().sum()
```

```python
events.drop_duplicates(inplace=True)
```

```python
df = pd.get_dummies(events['medal'])
```

```python
df = pd.concat([df,events],axis=1)
```

```python
df.groupby('noc').sum()[['Gold','Silver','Bronze']].sort_values(by='Gold',ascending=False).reset_index()
```
