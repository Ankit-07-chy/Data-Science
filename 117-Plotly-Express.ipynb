```python
# import datasets

df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv')
df.head()

tips = px.data.tips()
iris = px.data.iris()

gap = px.data.gapminder()

gap.head()
```

```python
# scatter plot using plotly go
x = gap[gap['year'] ==2007]

trace1 = go.Scatter(x=x['lifeExp'],y=x['gdpPercap'],mode='markers')
trace2 = go.Scatter(x=[0,1,333],y=[5,224,2],mode='lines')
data = [trace1,trace2]
layout = go.Layout(title='Life exp vs GDP per capita of 2007',xaxis={'title':'Life exp'},yaxis={'title':'Gdp'})
fig = go.Figure(data,layout)
fig.show()
```

```python
#plot life exp and gdp scater plot -> continent as color -> pop as size -> hover name -> range_x/range_y -> log_x/log_y


gap.head()
px.scatter(x,x='lifeExp',y='gdpPercap',color='continent',size='pop',size_max=50,hover_name='country')
```

```python
# plot animation of the above curve on the basic of year

gap.head()
px.scatter(gap,x='lifeExp',y='gdpPercap',color='continent',size='pop',size_max=50,hover_name='country',animation_frame='year',animation_group='country')
```

```python
#line plot
# plot india pop line plot

gap.head()
temp = gap[gap['country'] == 'India']
px.line(temp,x='year',y='pop',title='India Pop Growth')
```

```python
# 3d surface plot
#can not be created using plotly express
# we will use plotly gaph object->go

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx,yy = np.meshgrid(x,y)

z = xx**2 + yy**2

trace = go.Surface(x=x,y=y,z=z)
data=[trace]
layout=go.Layout(title='3D surface plot')
fig = go.Figure(data,layout)
fig.show()
```

```python
z = np.sin(xx)+np.cos(yy)

trace = go.Surface(x=x,y=y,z=z)
data=[trace]
layout=go.Layout(title='3D surface plot')
fig = go.Figure(data,layout)
fig.show()
```

```python
z = np.log(xx) + np.log(yy)
trace = go.Surface(x=x,y=y,z=z)
data=[trace]
layout=go.Layout(title='3D surface plot')
fig = go.Figure(data,layout)
fig.show()
```

```python
z = np.sinh(xx) + np.cosh(yy)
trace = go.Surface(x=x,y=y,z=z)
data=[trace]
layout=go.Layout(title='3D surface plot')
fig = go.Figure(data,layout)
fig.show()
```

```python
# contour plt

z = np.log(xx) + np.log(yy)
trace = go.Contour(x=x,y=y,z=z)
data=[trace]
layout=go.Layout(title='3D surface plot')
fig = go.Figure(data,layout)
fig.show()
```

```python
# subplots
from plotly.subplots import make_subplots
```

```python
fig = make_subplots(rows=1,cols=2)
fig.add_trace(
    go.Scatter(x=[1,9,4],y=[2,5,3],mode='markers'),
    row =1,
    col = 1
)

fig.add_trace(
    go.Bar(x=[1,9,4],y=[2,5,3]),
    row =1,
    col = 2
)
fig.update_layout(title='Subplt demo')
fig.show()
```
