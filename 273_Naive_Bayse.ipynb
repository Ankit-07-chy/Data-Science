```python
#just simple apply on that dataset than do for casestudy...
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd # Import pandas for get_dummies
from sklearn.naive_bayes import MultinomialNB
model1 = MultinomialNB(alpha=1.0)
from sklearn.naive_bayes import BernoulliNB
model2 = BernoulliNB(binarize=0.5)  # Threshold to convert features to binary
from sklearn.naive_bayes import CategoricalNB
model3 = CategoricalNB()#discrete categorical data
from sklearn.naive_bayes import ComplementNB
model4 = ComplementNB()#imbalanced text classification

# 1. Prepare your data
X=df.iloc[:,0:4]
y=df['Play Tennis']

# Convert categorical features to numerical using one-hot encoding
X = pd.get_dummies(X)

# 2. Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# 3. Create and train the model
model = GaussianNB()  # Or MultinomialNB/BernoulliNB
model.fit(X_train, y_train)

# 4. Make predictions
predictions = model.predict(X_test)

# 5. Evaluate performance
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy: {accuracy:.2f}")
```
